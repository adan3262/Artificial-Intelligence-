```python
import numpy
import matplotlib.pyplot as plt
# data
islamabad = [7, 9, 15, 22, 28, 33, 31, 30, 27, 21, 15, 9]
lahore = [10, 12, 18, 26, 33, 37, 36, 35, 31, 25, 19, 12]
karachi = [18, 19, 22, 26, 29, 32, 31, 31, 29, 27, 23, 20]
quetta = [5, 7, 12, 18, 24, 28, 27, 26, 23, 17, 11, 6]

# CreatING a NumPy array
temps = numpy.array([islamabad, lahore, karachi, quetta])

# Display the NumPy array
print("Numpy Array for all cities):")
print(temps)

# Calculating the average temperature
monthly_avg_temps = numpy.mean(temps, axis=0)

print("\nAverage temperature for each month across all cities:")
print("Jan\tFeb  Mar  Apr \tMay   June    July  Aug  Sep  Oct  Nov  Dec")
print(monthly_avg_temps)

months = numpy.arange(1, 13)  # January to December

# Line plot for each city
plt.figure(figsize=(12, 6))
plt.plot(months, islamabad, label="Islamabad")
plt.plot(months, lahore, label="Lahore")
plt.plot(months, karachi, label="Karachi")
plt.plot(months, quetta, label="Quetta")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.title("Monthly Temperatures for all the  Cities given")
plt.legend()

# Bar chart for average monthly temperatures
plt.figure(figsize=(8, 6))
plt.bar(months, monthly_avg_temps, color="skyblue")
plt.xlabel("Month")
plt.ylabel("Average Temperature (°C)")
plt.title("Average Monthly Temperature Across All Cities")
plt.xticks(months)

plt.show()
```

# QUESTION NO # 02

```python
# Define the hill's shape function
def hill_height(x):
    return -((x - 5) ** 2) + 20  # Peak at x = 5 with height 20

# Simulate hill climbing
def hill_climb(start_x):
    current_x = start_x
    while True:
        # Calculate height at current, left, and right positions
        current_height = hill_height(current_x)
        left_height = hill_height(current_x - 1)
        right_height = hill_height(current_x + 1)

        # Move if a neighboring point is higher
        if left_height > current_height:
            current_x -= 1
        elif right_height > current_height:
            current_x += 1
        else:
            break  # Stop when no higher point nearby

    return current_x, hill_height(current_x)

# Start at x = 2
start_x = 2
final_x, final_height = hill_climb(start_x)

print(f"Hasham and Mubarak reached the top at x = {final_x} with a height of {final_height}.")
```
